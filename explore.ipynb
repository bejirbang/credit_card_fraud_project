```python
#tahap modelling
cols_to_drop = [
    "sn", "first", "last", "street", "trans_num",
    "cc_num", "zip", "lat", "long",
    "merch_lat", "merch_long",
    "dob", "trans_date_trans_time", "unix_time",
    "city", "state", "gender", "job", "first", "last"
]

df_model = df.drop(columns=cols_to_drop)
```

```python
#one hot encoding 
merchant_fraud_rate = df.groupby("merchant")["is_fraud"].mean()
df_model["merchant_encoded"] = df_model["merchant"].map(merchant_fraud_rate)

category_fraud_rate = df.groupby("category")["is_fraud"].mean()
df_model["category_encoded"] = df_model["category"].map(category_fraud_rate)

df_model = df_model.drop(columns=["merchant", "category"])
```

```python
X = df_model.drop("is_fraud", axis=1)
y = df_model["is_fraud"]
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=200,
    class_weight="balanced",
    random_state=42
)

model.fit(X_train, y_train)
```

```python
y_proba = model.predict_proba(X_test)[:,1]

import numpy as np
y_pred_adjusted = np.where(y_proba > 0.2, 1, 0) 

from sklearn.metrics import classification_report, roc_auc_score
print(classification_report(y_test, y_pred_adjusted))
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
```

```python
# LOGISTIC REGRESSION MODELING

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.preprocessing import StandardScaler
import numpy as np


scaler_log = StandardScaler()

X_train_log = scaler_log.fit_transform(X_train)
X_test_log = scaler_log.transform(X_test)

log_model = LogisticRegression(class_weight='balanced', max_iter=1000)
log_model.fit(X_train_log, y_train)

y_proba_log = log_model.predict_proba(X_test_log)[:, 1]

y_pred_log = np.where(y_proba_log > 0.2, 1, 0)
```

```python
print(classification_report(y_test, y_pred_log))
print("ROC-AUC (Logistic):", roc_auc_score(y_test, y_proba_log))
```
